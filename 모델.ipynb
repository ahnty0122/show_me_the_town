```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data = pd.read_csv('./data.csv')
```

```python
print('구를 입력하세요: ')
x1 = str(input())
```

```python
test = pd.read_csv('./test.csv', encoding='CP949')
```

```python
print('인구밀도, 치안, 교육, 의료, 노인복지시설, 시장상점, 교통 점수를 입력하세요 : ')
want = pd.DataFrame(input().split()).transpose()
```

```python
want
```

```python
target = data[data['자치구']==x1]
target
```

```python
target2 = target.iloc[:, 2:]
```

```python
target2
```

```python
result = cosine_similarity(want, target2).transpose()
print(result)
```

```python
target['유사도'] = result
target
```

```python
final = pd.DataFrame(data=target['동'], columns=['동'])
```

```python
final['유사도'] = result
```

```python
final
```

```python
rank5 = final.sort_values(by='유사도', ascending=False)[:5]
rank5
```

```python
dongs = list(rank5['동'])
dongs
```

```python
final2 = target[target['동'].isin(dongs)].sort_values(by='유사도', ascending=False)
final2
```

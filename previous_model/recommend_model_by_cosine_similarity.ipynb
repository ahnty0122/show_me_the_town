```python
import pandas as pd
import numpy
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data = pd.read_csv('data.csv')
```

```python
print('구를 입력하세요: ')
x1 = str(input())
```

```python
print('인구밀도, 치안, 교육, 의료, 노인복지시설, 시장상점, 교통을 입력하세요: ')
want = pd.DataFrame(input().split()).transpose()
```

```python
target = data[data['자치구'] == x1]
```

```python
target
```

```python
target2 = target.iloc[:, 2:9]
```

```python
target2
```

```python
result = cosine_similarity(want , target2).transpose()
```

```python
final = pd.DataFrame(data = result, columns = ["similarity"], index = target['동'])
```

```python
final
```

```python
print(final.sort_values(by = 'similarity', ascending = False)[:5])
```
